```python

df = spark.table("workspace.mydatabase_db.fire_calls")
df = df.filter(df.CallType.isNotNull()) \
       .groupBy("CallType") \
       .count() \
       .withColumnRenamed("count", "call_count") \
       .orderBy("call_count", ascending=False) \
       .limit(10)
display(df)
```
