```python
from pyspark.sql.functions import lit
from pyspark.sql import Row

df = spark.createDataFrame([("Hello Spark", "Lets learn")], ["Column1", "Column2"])
df = df.withColumn("Column3", lit("Gen AI"))
df = df.withColumn("Column4", lit("Gen AI with Data bricks"))

new_row = Row(Column1="New Row", Column2="Added", Column3="Gen AI1", Column4="Gen AI2 with Data bricks")
df = df.union(spark.createDataFrame([new_row]))
new_row2 = Row(Column1="a", Column2="b", Column3="c", Column4="d")
df = df.union(spark.createDataFrame([new_row2]))
df = df.withColumn("Column5", lit("GPT 4"))
df.show()
```
