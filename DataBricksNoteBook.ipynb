```python
spark.sql("CREATE DATABASE IF NOT EXISTS mydatabase_db")
```

```python


spark.sql("""
    CREATE TABLE IF NOT EXISTS mydatabase_db.my_table (
        id INT,
        name STRING,
        age INT,
        city STRING,
        created_at TIMESTAMP
    )
""")

dummy_data = [
    (1, "Alice", 30, "New York", "2025-12-17 10:00:00"),
    (2, "Bob", 25, "Los Angeles", "2025-12-17 11:00:00"),
    (3, "Charlie", 35, "Chicago", "2025-12-17 12:00:00"),
    (4, "Diana", 28, "Houston", "2025-12-17 13:00:00"),
    (5, "Eve", 40, "Phoenix", "2025-12-17 14:00:00")
]

columns = ["id", "name", "age", "city", "created_at"]
df_dummy = spark.createDataFrame(dummy_data, columns)
df_dummy.write.mode("overwrite").insertInto("mydatabase_db.my_table")
```

```python
df_read = spark.sql("SELECT * FROM mydatabase_db.my_table")
display(df_read)
```

4. Create file_call table from  a file

```python

df_fire_calls = spark.read.option("header", True).option("inferSchema", True).csv("/Volumes/workspace/mydatabase_db/Files/sf-fire-calls.csv")
df_fire_calls.write.mode("overwrite").saveAsTable("mydatabase_db.fire_calls")
```

```python
row_count = spark.table("mydatabase_db.fire_calls").count()
display(row_count)
```
